#Data Loading

```python
from tensorflow.keras import utils
import numpy as np
import matplotlib.pyplot as plt
```

```python
img=utils.load_img("images/Train/Pikachu/297.jpg", target_size=(100,100)) #diff size problem now resolved
img
```

```python
img=np.array(img)
img.shape
```

```python
train_data=[] #X
train_labels=[] #Y

for category in classes:

  folder="images/Train/" + category
  for img_name in os.listdir(folder):
      img_path=folder + "/" + img_name

      img=utils.load_img(img_path, target_size=(100,100))
      img=np.array(img)

      train_data.append(img)
      train_labels.append(category)
```

```python
X_train=np.array(train_data)
Y_train=np.array(train_labels)
```

```python
X_train.shape
```

```python
Y_train.shape
```

```python
plt.imshow(X_train[20])
```

```python
Y_train[20]
```

```python
Y_train
```

```python
category2label={"Pikachu" : 0, "Bulbasaur" : 1, "Charmander" : 2}
label2category={0: "Pikachu", 1 : "Bulbasaur", 2 : "Charmander"}
```

```python
Y_train=np.array([category2label[label] for label in Y_train])
```

```python
Y_train[:5]
```

```python
from tensorflow.keras.utils import to_categorical
```

```python
Y_train.shape
```

```python
Y_train_new=to_categorical(Y_train)
Y_train_new.shape
```

```python
Y_train_new[:5]
```

```python
from tensorflow.keras import Sequential #we want to create sequential model
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Convolution2D, MaxPooling2D, Flatten
```

```python
model=Sequential()

model.add(Convolution2D(32,(3,3), activation="relu", input_shape=(100,100,3)))
model.add(Convolution2D(64,(3,3), activation="relu"))
model.add(MaxPooling2D())
model.add(Convolution2D(128,(3,3), activation="relu"))
model.add(MaxPooling2D())
model.add(Flatten())
model.add(Dense(3,activation="softmax"))
```

```python
model.compile(optimizer="adam", loss="categorical_crossentropy",metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(X_train,Y_train_new,batch_size=64,epochs=25)
```

#Test Data

```python
test_data=[] #X
test_labels=[] #Y

for category in classes:

  folder="images/Test/" + category
  for img_name in os.listdir(folder):
      img_path=folder + "/" + img_name

      img=utils.load_img(img_path, target_size=(100,100))
      img=np.array(img)

      test_data.append(img)
      test_labels.append(category)
```

```python
X_test=np.array(test_data)
Y_test=np.array(test_labels)
```
